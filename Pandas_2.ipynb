# iloc vs Loc
1. iloc

* iloc e index number diye fetch krete hy
2. Loc

* loc te index name diye fetch krte hoy

```python
student_dict = {
    'name': ['ayan', 'hena', 'sohan', 'mayan', 'helal', 'siyam'],
    'id': [100, 90, 80, 120, 0, 0],
 'iq': [80, 70, 50, 100, 0, 0],
 'marks': [10, 7, 2, 14, 0, 0]\
}
students = pd.DataFrame(student_dict)
students.set_index('name', inplace= True)
```

```python
# single row
movies.iloc[0] # first row starts from 0
```

```python
# loc
print(students)
students.loc['ayan'] #fetch specific data from index
students.loc['ayan':'mayan'] #fetch data by selecting range from index
students.loc['ayan': 'siyam':2] #fetch data by selecting range from index with step


# fancy indexing
students.loc[['ayan', 'siyam', 'sohan', 'helal']] # for fancy indexing, there will be [[---]]
```

```python
# how many matches CSK won at Kolkata

# csk_won = ipl['WinningTeam'] == 'Chennai Super Kings'
# venue = ipl['Venue'] == 'Eden Gardens'
# data = ipl[venue & csk_won]
# # data.shape[0]
# data.shape[0]

def team_result_at_venue(venue, team):
  mask1 = ipl['Venue'] == venue
  mask2 = ipl['WinningTeam'] == team
  data = ipl[mask1 & mask2]
  return data.shape[0]

team_result_at_venue('Wankhede Stadium', 'Rajasthan Royals')
```
